```python
import javalang
import os
import pandas as pd
import re
import sys

def extract_method_metrics (methods):
    
    for method in methods :
    
        for n, m in method:
            if(type(m).__base__ ==javalang.tree.Statement and not (type(m) == javalang.tree.BlockStatement)):
                s += 1
            if (type(m) in [javalang.tree.IfStatement, javalang.tree.WhileStatement, javalang.tree.DoStatement]):
                c +=1
            if(type(m) == javalang.tree.MethodDeclaration and not m.throws == None):
                e += len(m.throws)
        szs.append(s)
        cpxs.append(c) 
        errs.append(e)
        rets.append(r)
    return max(szs), max(cpxs), max(errs), max(rets)



def extract_nlp_metrics(node):
    bcm,nml,wrd,dcm = 0,0,0,0
    for m,n in node:
        if(hasattr(n,"documentation") and n.documentation):
            bcm+=1
            words = re.findall("\w+", n.documentation)
            wrd += len(words)
    statement_count =0
    m_name_len = 0
    if node.methods is not None:
        size = len(node.methods)
        for method in node.methods:
            m_name_len += len(method.name)
            for path, node in method.filter(javalang.tree.Statement):
                if not isinstance(node, javalang.tree.BlockStatement):
                    statement_count +=1;
        nml = m_name_len / size if size > 0 else 0
    if(statement_count == 0): 
        dcm =0
    else :
        dcm = wrd/statement_count
    return bcm,nml,wrd,dcm


def main(path):
#     path = sys.argv[1]
    
    class_names = []
    #method count
    MTH= []
    # field count
    FLD= []
    # public called methods count
    RFC = [ ]
    # implemented interfaces count
    INT= []
    # statements count
    SZ= []
    # conditional loop statements count
    CPX= []
    # exceptions in throws clause count
    EX= []
    # return points count
    RET = [ ]
    # block comments count
    BCM= []
    # avg length of method names
    NML= [ ]
    # words in blocks count
    WRD= []
    # words per statement count
    DCM = [ ]

    for root , dirs , files in os.walk(path):
        for name in files:
            if name.endswith(".java"):
                str = open(os.path.join(root, name), "r").read()
                tree = javalang.parse.parse(str)
                print (name)
                className = name.replace(".java","")
                for path , node in tree.filter(javalang.tree.ClassDeclaration ):
                    #get the class one 
                    if(node.name == className):
                        class_names.append(node.name)
                        MTH.append(len(node.methods))
                        FLD.append(len(node.fields)) 
                        if (node.implements == None):
                            INT.append(0)
                        else :
                            INT.append(len(node.implements)) 

                        n_of_rfc=0
                        for method in node.methods:
                            if ("public" in method.modifiers):
                                n_of_rfc +=1
                            n_of_rfc += len(list(method.filter(javalang.tree.MethodInvocation)))
                        RFC.append(n_of_rfc)
                        ## EXTRACT METHOD METRICS
                        sz , cpx , ex , ret = extract_method_metrics(node.methods)
                        SZ.append(sz)
                        CPX.append(cpx)
                        EX.append(ex) 
                        RET.append(ret)
                        ## EXTRACT NLP METRICS
                        bcm, nml, wrd, dcm = extract_nlp_metrics(node)
                        BCM.append(bcm)
                        NML.append(nml)
                        WRD.append(wrd)
                        DCM.append(dcm)
#     df = pd.DataFrame({"class name": class_names, "MTH": MTH, "FLD": FLD, "RFC":RFC, "INT": INT,"SZ":SZ, "WRD" :WRD, "DCM" :DCM})
    print(df)

main("android-chess/app/src/main/java/jwtc/chess/")
```
